# Exercício - Limpeza e Tratamento de Dados

```python
#Por Judah Benjamin

#Criação do DataFrame:
"""
Crie um DataFrame chamado df_vendas com as seguintes colunas: produto, preco, quantidade, e ano_venda.
Insira pelo menos 10 registros, sendo que alguns dos valores em preco e ano_venda
devem ser nulos (use np.nan para representar os valores nulos).
"""


import numpy as np
import pandas as pd

df_vendas = pd.DataFrame({
    'produto': [
        'Arroz',
        'Manteiga',
        'Feijão',
        'Óleo de Soja',
        'Açúcar',
        'Café',
        'Leite',
        'Pão',
        'Ovos',
        'Sabão em Pó'
    ],
    'preco': [
        5.50, 12.00, 8.75, 7.99, np.nan,
        15.00, 4.20, 6.50, 10.00, np.nan
    ],
    'quantidade': [
        100, 50, 75, np.nan, 120,
        90, 200, 60, np.nan, 80
    ],
    'ano_venda': [
        2021, 2022, np.nan, 2023, 2021,
        2022, 2023, 2021, np.nan, 2022
    ]
})

display(df_vendas) #Comando para exibir DataFrames(e outras estruturas de dados) de forma mais bonita e interativa.
#Representação formatada da tabela para uma melhor visualização dos dados.
```

```python
#Identificação de Dados Nulos:
"""
Utilize o método isnull() para identificar quais colunas têm valores nulos e quantos são.
"""

df_vendas.isnull()
```

```python
df_vendas.isnull().sum()
#Soma tudo o que está nulo
```

```python
#Tratamento de Dados Nulos:
"""
Preencha os valores nulos na coluna preco com a média dos preços.
"""
# Ajuste: Preenchendo a coluna 'preco' diretamente com a média, como solicitado.
df_vendas['preco'] = df_vendas['preco'].fillna(df_vendas['preco'].mean().round(2))

"""
Preencha os valores nulos na coluna ano_venda com o valor "Não informado".
"""
# Ajuste: Preenchendo a coluna 'ano_venda' diretamente com 'Não informado', como solicitado.
df_vendas['ano_venda'] = df_vendas['ano_venda'].fillna('Não informado')

# Exibe o DataFrame após o tratamento dos nulos
df_vendas
```

```python
#Remoção de Registros:
"""
Se houver algum registro onde a quantidade seja igual a 0, remova esses registros do DataFrame.
"""

# Ajuste: Primeiro, removemos as linhas onde 'quantidade' é nula para garantir que a conversão para int seja possível.
# Isso também ajuda a lidar com registros inconsistentes antes de verificar o valor 0.
df_vendas = df_vendas.dropna(subset=['quantidade'])

# Ajuste: Em seguida, filtramos e removemos apenas os registros onde a 'quantidade' é igual a 0,
# seguindo a instrução específica do exercício.
df_vendas = df_vendas[df_vendas['quantidade'] != 0]

# Verifica novamente a soma de nulos após as remoções para confirmar a limpeza.
df_vendas.isnull().sum()
# Obs: você também pode especificar a coluna com dropna(subset=['nome da coluna']).
```

```python
#Conversão de Tipos:
"""
Verifique se a coluna ano_venda está como string e converta-a para um tipo inteiro, se necessário.
"""

# Ajuste: Converte a coluna 'quantidade' para o tipo inteiro.
# Esta conversão é segura agora, pois os valores nulos e zero foram tratados na etapa anterior.
df_vendas = df_vendas.assign(quantidade=df_vendas['quantidade'].astype(int))

# Observação: A instrução pedia para converter 'ano_venda' para inteiro.
# No entanto, a coluna 'ano_venda' agora contém strings ('Não informado') que resultaram do preenchimento de nulos.
# Portanto, não é possível convertê-la diretamente para um tipo inteiro sem antes remover ou transformar essas strings.
# A coluna 'ano_venda' permanecerá como tipo 'object' (string) devido a essa incompatibilidade de tipos.
# Se a intenção final for ter 'ano_venda' como int, o tratamento dos valores nulos deveria ser diferente
# (ex: preencher com um valor numérico como 0, ou remover as linhas com NaN).

# Exibe as primeiras linhas do DataFrame com os tipos atualizados.
df_vendas.head()
```
