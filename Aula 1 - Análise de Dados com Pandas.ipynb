```python
import pandas as pd

# Recarregar o DataFrame com o separador correto
df = pd.read_csv("http://raw.githubusercontent.com/contaratoandressa/movie_analysis/refs/heads/master/data_model.csv", sep=';')

# Renomear as colunas para nomes mais compreensíveis
column_mapping = {
    'data': 'ano',
    'genero': 'genero_filme',
    'ingresso': 'total_ingresso_vendido',
    'cinema': 'receita_bruta_cinema',
    'google_trends': 'score_google_trends',
    'evn': 'score_evn',
    'prop_cinema': 'propriedade_cinema'
}

df = df.rename(columns=column_mapping)

# Exibir as primeiras linhas e as novas colunas para verificar
print("DataFrame com colunas renomeadas:")
print(df.head())
print("\nNovos nomes das colunas:")
print(df.columns)
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'])
print("Coluna 'Unnamed: 0' removida.")
print(df.head())
```
